```python
#Store Customers Orders
#Store each customer's order details (customer name, product, price, category) as tuples inside a list
l=[("neerja","nike",34.4,"shoes"),("rishabh","pizza",47.4,"food"),("arjun","almirah",107.78,"furniture"),("riya","iphone",56,"electronics"),("rishabh","laptop",67,"electronics"),("dhruv","pillows",78.6,"cushion"),("neerja","chair",64.4,"furniture"),("arjun","jockey",92.2,"clothes"),("rishabh","pen",52.1,"stationary"),("riya","skirt",31,"clothes"),("arjun","ipad",12,"electronics")]
```

```python
#Analyze customer orders
# Use a loop to calculate the total amount each customer spends
w=set()
s={}
for h in range(length):
  w.add(l[h][0])
r=list(w)
print(r)
for f in range(len(r)):
  n=0
  for g in range(length):
    if r[f]==l[g][0]:
     n+=l[g][2]
  s[r[f]]=n
print(s)
```

```python
# If the total purchase value is above $100, classify the customer as a high-value buyer
# If it is between $50 and $100, classify the customer as a moderate buyer
# If it is below $50, classify them as a low-value buyer
t={}
for key in s:
  if s[key]>=100:
    t[key]="high-value buyer"
  elif 50<=s[key]<100:
    t[key]="moderate-value buyer"
  else:
    t[key]="low-value buyer"
print(t)
```
